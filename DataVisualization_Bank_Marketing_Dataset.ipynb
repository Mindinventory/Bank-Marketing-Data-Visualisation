# **About bank.csv file**

bank.csv is the classic marketing bank dataset. The dataset gives you information about a marketing campaign of a financial institution in which you will have to analyze in order to find ways to look for future strategies in order to improve future marketing campaigns for the bank.

# **Key Points:**

**Data Source:** The code analyzes data from a file named "*bank.csv*"

**Feature Types:** The data contains both numerical and categorical features.

```python
import pandas as pd
from pandas.plotting import parallel_coordinates
df=pd.read_csv("bank.csv")
```

# Spliting dataset into numerical features and catagorical features for visualisation

```python
numerical_features=df.select_dtypes(include='number')
categorical_features=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
```

# Finding distribution of categorical feature

# **CountPlot**

# **Summary of CountPlot**

**Interpreting Count Plots:**

*   X-axis: Represents the number of instances (counts) for each category.
*   Y-axis: Represents the different categories within the feature.
*   Bar Height: Indicates the frequency of each category.
*   Comparisons: Count plots enable you to visually compare the relative frequencies of different categories within a feature.

**Considerations from each graph**

1.   Job:
  *   The most common job type is "management" followed by "blue-collar"
  *   Less frequent job types include "entrepreneur", "housemaid", "self-employed".

2.   Maritial Status:
  *   The most common marital status is "married".
  *   Less frequent marital statuses include "divorced".

3.   Education:
  *   The most common education level is "secondary" followed by "tertiary".
  *   Less frequent education levels include "primary" and "unknown".

4.   Default:
  *   The majority of customers (around 98%) do not have a default.
  *   A small percentage (around 2%) have a default.

5.   Housing:
  *   Frequency of customers having housing and not having housing is almost 50-50%

6.   Loan:
  *   The majority of customers (around 85%) do not have a loan.
  *   A small percentage (around 15%) have a loan.

7.   Contact:
  *   The most common contact method is "cellular".
  *   Less frequent contact methods include "telephone" followed by "unknown."

8.   Month:
  *   "May" month seems having a significantly higher frequency than other months.
  *   "December" month seems having a significantly higher frequency than other months.

9.   Poutcome:
  *   The most common outcome of the Poutcome marketing campaign is "Unknown," followed by "failuer" and "Success".

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,80))
plotnumber =1
for categorical_feature in categorical_features:
    ax = plt.subplot(12,3,plotnumber)
    sns.countplot(y=categorical_feature,data=df)
    plt.xlabel(categorical_feature)
    plt.title(categorical_feature)
    plotnumber+=1
plt.show()
```

# **Categorical plot (catplot)**

#**Summary of Categorical plot (catplot)**

**Interpreting catplot:**


* Type: Categorical plot (catplot), a grid of bar charts designed to visualize relationships between a categorical variable and a target variable.
* X-axis: Target Variable - Deposit (yes/no), indicating whether an individual opened a deposit.
* Y-axis: Categorical Variable.
* Organization: The plot is divided into subplots, one for each data of a particular category.
* Bars: Represent the counts of individuals with "yes" or "no" deposit within each

**Considerations from all graphs of all categories**

* Count of each data of each category is displayed
* Let's take example of the catplot graph for the "marital" category:

# Finding distribution of numerical feature



# **Histogram**

#**Summary of Histogram**

**Interpreting Histogram:**
* Type: Histogram, a bar graph that visualizes the distribution of numerical data.
* X-axis: Numerical Feature values.
* Y-axis: Frequency Range.
* Bars: Represent the frequency of individuals within each Numerical Feature.

**Considerations from all Histogram graphs**
*   Frequency distribution of each numerical feature is displayed.
*   Let's take example of Distribution of age:
  * Distribution Shape: Right-skewed, meaning more individuals are concentrated in younger age ranges, with a longer tail towards older ages.
  * Peak: The highest frequency occurs in the 30-40 age range, indicating the most common age group.
  * Tail: The distribution gradually tapers off towards older ages, with fewer individuals in the 60-80 range.
  * Notable Features:A slight dip in frequency around the 50-60 age range.


```python
for nf in numerical_features:
    plt.figure(figsize=(3, 2))
    plt.hist(numerical_features[nf], color='#c6ac8f')
    plt.title(f'Distribution of {nf}')
    plt.xlabel(nf)
    plt.ylabel('Frequency')
    plt.grid(False)
    plt.show()
```

# **Density plot**

#**Summary of Density Plot also known as Kernel Density Estimate (KDE)**

**Interpreting Density Plot:**
* Type: Density plot, also known as kernel density estimate (KDE), visualizes the distribution of numerical data using a smooth curve.
* X-axis: Numerical Feature values.
* Y-axis: Density (not frequency), indicating the probability of a data point occurring at a given numerical feature.
* Curve: Represents the estimated density of the numerical feature distribution.

**Considerations from all Density Plot**
*   Density distribution of each numerical feature is displayed.
*   Let's take example of Density of age:
  * Right-skewed: Similar to the histogram, the curve leans to the right, suggesting a concentration of individuals in younger ages and a longer tail towards older ages.
  * Peak: The highest density occurs around 35-40 years, indicating the most common age range.
  * Tail: The curve gradually declines towards older ages, with lower density values for those above 60.


```python
for nf in numerical_features:
    plt.figure(figsize=(6, 4))
    sns.kdeplot(data=df, x=nf, fill=True)
    plt.title(f'Density Plot of {nf}')
    plt.xlabel(nf)
    plt.ylabel('Density')
    plt.show()
```

# **Violin Plot**

#**Summary of Violin**

**Interpreting Violin Plot:**
* Type: Violin plot, a combination of a box plot and a density plot, providing a comprehensive view of the distribution of numerical data.
* X-axis: Numerical Feature values.
* Y-axis: values of numerical features
* Shape: A vertical, symmetrical shape resembling a violin, with a white dot representing the median numerical value.

**Considerations from all Violin Plot**
*   Violin Plot of each numerical feature is displayed.
*   Let's take example of Violin of age:
  * Right-skewed: The violin's shape is wider towards the right, confirming the right-skewed distribution observed in previous plots.
  * Peak: The highest density region is around 35-40 years, aligning with the peak in the density plot. The white dot within the box plot indicates the median age, likely in the mid-30s
  * Tail: The density gradually decreases towards older ages, forming the longer tail on the right side of the violin

```python
for nf in numerical_features:
    plt.figure(figsize=(6, 4))
    sns.violinplot(x=nf, data=df)
    plt.title(f'Violin Plot of {nf}')
    plt.xlabel(nf)
    plt.ylabel('Value')
    plt.show()
```

# **Box plot**

#**Summary of BoxPlot**

**Interpreting BoxPlot:**
* Type: Box plot, visualizing the distribution of numerical data, focusing on quartiles and potential outliers.
* X-axis: Numerical Feature values.
* Box: Represents the middle 50% of the data, extending from the first quartile (Q1) to the third quartile (Q3).
* Median: Line within the box marking the 50th percentile (Q2), the middle value in the dataset.
* Outliers: Points beyond the whiskers, potentially unusual values.

**Considerations from all BoxPlot**
*   BoxPlot of each numerical feature is displayed.
*   Let's take example of BoxPlot of age:
  * Right-skewed: The box is closer to the left side of the plot, with a longer whisker on the right, indicating a concentration of values towards the lower end and a tail of higher values.
  * Median: The median line within the box is approximately around 35-40 years, suggesting the middle value in the dataset.

```python
for coloum in numerical_features:
    sns.boxplot(data=df, x=coloum)
    plt.xlabel(coloum)
    plt.title('Box Plot of '+coloum)
    plt.grid(False)
    plt.show()
```

* X-axis: Deposit status (yes/no).
* Y-axis: Mean value of the numerical feature.
* Bars: Represent the mean value for each group, with error bars indicating variability.

# **HeatMap**

#**Summary of HeatMap**

**Interpreting HeatMap:**

* Type: Correlation heatmap, a visual representation of the correlation coefficients between pairs of numerical variables.
* Size: 7x7, indicating 7 numerical features analyzed.
* Color Gradient: Red (strong positive correlation) -> Blue (strong negative correlation).
* Numbers: Decimal values within each cell, representing the correlation coefficients.

**Considerations of Heatmap**
* Color Intensity: Stronger correlations are indicated by more intense colors (darker red or blue).
* Positive Correlations: Cells with red signify variables that tend to increase or decrease together.
  * Example: age - age (1.00).
* Negative Correlations: Cells with blue signify variables that tend to move in opposite directions.
  * Example: day - pday (-0.08).
* No Correlation: Cells close to white indicate little or no correlation
  * Example: pdays - previous (0.51).

```python
correlation_matrix =numerical_features.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

# Multivariate Analysis

# **Pairplot**

#**Summary of Pairplot**

**Interpreting Pairplot:**
* Type: Pairplot, a grid of scatter plots visualizing relationships between pairs of numerical variables.
* Diagonal: Histograms of each variable's distribution.
* Upper Triangle: Scatter plots with variable 1 on the x-axis and variable 2 on the y-axis.
* Lower Triangle: Mirrored scatter plots with variables reversed.

**Considerations of Pairplot**
* Identify Strong Correlations: Note pairs of variables with clear linear relationships.
  * In the scatter plot of "balance" vs. "duration,"there is a tight, linear upward trend, suggesting a strong positive correlation between these variables. This means that individuals with higher balances tend to have longer durations.
* Explore Non-Linear Relationships: Investigate curves or patterns that suggest more complex associations.
 * In the scatter plot of "age" vs. "balance," there is a straight line, which indicates a non-linerar relationship. This means that the association between age and balance changes over different age ranges.
* The diagonal of a pairplot consists of a series of histograms, positioned along the main diagonal from the top left to the bottom right. These histograms visualize the distribution of each individual numerical variable within the dataset.
  * Imagine a diagonal histogram for "balance" that is right-skewed, centered around 500, and with a range from 0 to 2500. This suggests:
  Most individuals have balances around 500.
  A few individuals have much higher balances, up to 2500.
  The data is not normally distributed.

```python
sns.pairplot(numerical_features)
plt.show()
```

# **Parallel Coordinate**

#**Summary of Parallel Coordinate**

**Interpreting Parallel Coordinate:**
* Type: Parallel coordinates plot, visualizing multivariate numerical data by plotting each observation as a line across multiple axes.
* X-axis: Each vertical axis represents one numerical variable.
* Y-axis: Range of days
* Lines: Each line represents a single observation (individual or data point).
* Color Gradient: The color of the lines varies based on the "day" variable (viridis colormap), indicating how values for other variables change across different days.

**Considerations of Parallel Coordinate**
* There appears to be a positive correlation between age and balance.
* There appears to be a negative correlation between age and duration.


```python
plt.figure(figsize=(10, 6))
parallel_coordinates(numerical_features,'day', colormap='viridis')
plt.title('Parallel Coordinates Plot of Numerical Columns')
plt.show()
```

#**Contour plot**

**Interpreting Contour plot:**
* Type: Parallel coordinates plot, visualizing multivariate numerical data by plotting each observation as a line across multiple axes.
* Type: Contour plot, visualizing a 3-dimensional relationship between three numerical variables using a 2-dimensional surface.
* X-axis: Age (years)
* Y-axis: Balance
* Color Gradient: Fill colors represent the third variable, "duration," using the viridis colormap (blue to yellow, indicating lower to higher values).
Contour Lines: Black lines outline regions of similar "duration" values, like elevation lines on a topographic map.

**Considerations of Contour plot**
* Identify high and low regions: Look for areas with consistently high or low "duration" values based on color.
  * Example: A dark blue region might suggest low "duration" for combinations of low "age" and high "balance."
* Examine trends and patterns: Observe how colors and contour lines change across the plot.
  * Example: A diagonal trend from blue to yellow might indicate increasing "duration" with increasing "age" and "balance."

```python
x_column = 'age'
y_column = 'balance'
z_column = 'duration'

plt.figure(figsize=(6, 4))
contour = plt.tricontourf(df[x_column], df[y_column], df[z_column], cmap='viridis')
plt.colorbar(contour)
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title('Contour Plot')
plt.show()
```
