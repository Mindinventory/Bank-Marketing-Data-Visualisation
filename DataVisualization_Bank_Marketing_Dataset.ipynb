```python
import pandas as pd
import numpy as np
from pandas.plotting import parallel_coordinates
df=pd.read_csv("bank.csv")
```

# Spliting dataset into numerical features and catagorical features for visualisation

```python
numerical_features=df.select_dtypes(include='number')
categorical_features=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
```

# Finding distribution of categorical feature

CountPlot

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,80))
plotnumber =1
for categorical_feature in categorical_features:
    ax = plt.subplot(12,3,plotnumber)
    sns.countplot(y=categorical_feature,data=df)
    plt.xlabel(categorical_feature)
    plt.title(categorical_feature)
    plotnumber+=1
plt.show()
```

Frequency polygon

```python
for nf in categorical_features:
    plt.figure(figsize=(6, 4))
    freq = df[nf].value_counts().sort_index()
    plt.plot(freq.index, freq.values, marker='o', linestyle='-')
    plt.title(f'Frequency Polygon of {nf}')
    plt.xlabel(nf)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.show()
```

# Relationship between Categorical Features and Label

```python
col_wrap_value = len(categorical_features)
for categorical_feature in categorical_features:
    sns.catplot(x='deposit', col=categorical_feature, kind='count', data= df, col_wrap=col_wrap_value)
plt.show()
```

# Finding distribution of numerical feature

Histogram

```python
for nf in numerical_features:
    plt.figure(figsize=(3, 2))
    plt.hist(numerical_features[nf], color='#c6ac8f')
    plt.title(f'Dstribution of {nf}')
    plt.xlabel(nf)
    plt.ylabel('Frequency')
    plt.grid(False)
    plt.show()
```

Density plot

```python
for nf in numerical_features:
    plt.figure(figsize=(6, 4))
    sns.kdeplot(data=df, x=nf, fill=True)
    plt.title(f'Density Plot of {nf}')
    plt.xlabel(nf)
    plt.ylabel('Density')
    plt.show()
```

Violin Plot

```python
for nf in numerical_features:
    plt.figure(figsize=(6, 4))
    sns.violinplot(x=nf, data=df)
    plt.title(f'Violin Plot of {nf}')
    plt.xlabel(nf)
    plt.ylabel('Value')
    plt.show()
```

BarGraph

```python
for nf in numerical_features:
    plt.figure(figsize=(3, 2))
    sns.barplot(x='deposit', y=numerical_features[nf], data=df)
    plt.title(f'{nf} vs deposit')
    plt.show()
```

HeatMap

```python
correlation_matrix =numerical_features.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

Box plot box and whisker plot

```python
for coloum in numerical_features:
    sns.boxplot(data=df, x=coloum)
    plt.xlabel(coloum)
    plt.title('Box Plot of '+coloum)
    plt.grid(False)
    plt.show()
```

# Multivariate Analysis

Scatter plot

```python
sns.pairplot(numerical_features)
plt.show()
```

Parallel Coordinate

```python
plt.figure(figsize=(10, 6))
parallel_coordinates(numerical_features,'day', colormap='viridis')
plt.title('Parallel Coordinates Plot of Numerical Columns')
plt.show()
```

 Contour plot

```python
x_column = 'age'
y_column = 'balance'
z_column = 'duration'

plt.figure(figsize=(6, 4))
contour = plt.tricontourf(df[x_column], df[y_column], df[z_column], cmap='viridis')
plt.colorbar(contour)
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title('Contour Plot')
plt.show()
```
